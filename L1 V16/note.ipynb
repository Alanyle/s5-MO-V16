## Сводная информация по датафрейму

### Subtask:
Получить сводную информацию о загруженном DataFrame, используя методы, такие как .info(), .describe(), .head().


**Обоснование**:
Нужно вывести сводную информацию о датафрейме с помощью методов `.info()`, `.describe()` и `.head()`. Удобно выполнить эти вызовы в одном кодовом блоке.



```python
print("Сводка DataFrame:")
df.info()
print("\nОписательные статистики:")
print(df.describe())
print("\nПервые 5 строк (повторно):")
print(df.head())
```

**Обоснование**:
Следующий шаг — выбрать и вывести столбец `Goals` из датафрейма. Для этого достаточно стандартного обращения к столбцу pandas.



```python
print("Столбец Goals:")
print(df['Goals'])
```

**Обоснование**:
Теперь нужно посчитать общее число уникальных команд. Для этого воспользуюсь столбцом `Team` и методом `.nunique()`.



```python
unique_teams_count = df['Team'].nunique()
print(f"Общее число уникальных команд: {unique_teams_count}")
```

**Обоснование**:
Далее нужно определить количество строк и столбцов в наборе данных. Это удобно сделать через атрибут `.shape` датафрейма.



```python
num_rows, num_cols = df.shape
print(f"Количество строк: {num_rows}")
print(f"Количество столбцов: {num_cols}")
```

**Обоснование**:
Теперь необходимо создать новый датафрейм `discipline`, включающий столбцы `Team`, `Yellow Cards` и `Red Cards`. Просто выберу эти колонки из исходного `df`.



```python
discipline = df[['Team', 'Yellow Cards', 'Red Cards']]
print("Датафрейм discipline создан. Первые 5 строк:")
print(discipline.head())
```

**Обоснование**:
Далее нужно отсортировать датафрейм `discipline` по столбцам `Red Cards`, затем `Yellow Cards` по убыванию. Для этого подойдёт метод `.sort_values()`.



```python
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("discipline, отсортированный по Red Cards и Yellow Cards (по убыванию). Первые 5 строк:")
print(discipline_sorted.head())
```

**Обоснование**:
Нужно вычислить среднее количество жёлтых карточек на команду. Возьму столбец `Yellow Cards` и применю метод `.mean()`.



```python
average_yellow_cards = df['Yellow Cards'].mean()
print(f"Среднее число жёлтых карточек на команду: {average_yellow_cards:.2f}")
```

**Обоснование**:
Далее требуется вывести команды, забившие более 6 голов. Для этого отфильтрую датафрейм `df` по условию для столбца `Goals`.



```python
teams_more_than_6_goals = df[df['Goals'] > 6]
print("Команды, забившие более 6 голов:")
print(teams_more_than_6_goals[['Team', 'Goals']])
```

**Обоснование**:
Затем нужно вывести данные команд, названия которых начинаются на букву «G». Использую строковый метод `str.startswith` для фильтрации по столбцу `Team`.



```python
teams_starting_with_G = df[df['Team'].str.startswith('G')]
print("Команды, начинающиеся с буквы 'G':")
print(teams_starting_with_G[['Team', 'Goals']])
```

**Обоснование**:
Теперь необходимо выбрать и показать первые 7 столбцов исходного датафрейма. Для этого воспользуюсь позиционным индексированием `iloc`.



```python
first_7_columns = df.iloc[:, :7]
print("Первые 7 столбцов DataFrame:")
print(first_7_columns.head())
```

**Обоснование**:
Затем нужно вывести все столбцы, кроме последних трёх. Это также удобно сделать с помощью `iloc`, выбрав все строки и колонки до третьей с конца.



```python
columns_except_last_3 = df.iloc[:, :-3]
print("Все столбцы, кроме последних трёх:")
print(columns_except_last_3.head())
```

**Обоснование**:
Далее требуется определить точность ударов по воротам для Англии, Италии и России. Сначала преобразую столбец `Shooting Accuracy` в числовой формат, убрав символ `%` и приведя значения к `float`, затем отфильтрую нужные команды и выведу их показатели.



```python
df['Shooting Accuracy'] = df['Shooting Accuracy'].str.replace('%', '').astype(float)

selected_countries = ['England', 'Italy', 'Russia']
shooting_accuracy_selected = df[df['Team'].isin(selected_countries)][['Team', 'Shooting Accuracy']]

print("Точность ударов Англии, Италии и России:")
print(shooting_accuracy_selected)
```

**Обоснование**:
Теперь нужно определить TOP‑5 команд с наибольшим и наименьшим количеством нарушений. Сначала отсортирую датафрейм по `Fouls Conceded` по убыванию, чтобы получить лидеров по фолам.



```python
top_5_most_fouls = df.sort_values(by='Fouls Conceded', ascending=False).head(5)
print("TOP-5 команд с наибольшим количеством фолов:")
print(top_5_most_fouls[['Team', 'Fouls Conceded']])
```

**Обоснование**:
После определения лидеров по нарушениям нужно вывести и команды с минимальным количеством фолов. Для этого отсортирую `df` по `Fouls Conceded` по возрастанию и возьму первые пять строк.



```python
top_5_least_fouls = df.sort_values(by='Fouls Conceded', ascending=True).head(5)
print("TOP-5 команд с наименьшим количеством фолов:")
print(top_5_least_fouls[['Team', 'Fouls Conceded']])
```
