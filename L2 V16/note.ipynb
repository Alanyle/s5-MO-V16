```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

sns.set_theme(style="whitegrid", palette="crest")
```

```python
data_path = "onlinefoods.csv"

df = (
    pd.read_csv(data_path)
      .dropna(axis=1, how='all')
)
df.columns = (
    df.columns
      .str.strip()
      .str.replace(' ', '_')
)
for col in ['Output', 'Feedback', 'Monthly_Income', 'Marital_Status', 'Occupation', 'Educational_Qualifications']:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

income_order = ['No Income', 'Below Rs.10000', '10001 to 25000', '25001 to 50000', 'More than 50000']
df['Monthly_Income'] = pd.Categorical(df['Monthly_Income'], categories=income_order, ordered=True)

df['Pin_code'] = df['Pin_code'].astype(str)

print(f"Файл загружен: {data_path}. Размер таблицы: {df.shape[0]} строк x {df.shape[1]} столбцов")
df.head()
```

# Онлайн-заказ еды: первичный анализ

 Вариант 16 – Онлайн-заказ еды
Датасет – onlinefoods.csv
Набор данных содержит информацию, собранную с онлайн-платформы заказа
еды за определенный период времени. Он охватывает различные атрибуты, связанные
с профессией, размером семьи, отзывами и т.д.
Атрибуты:
Демографическая информация:
• Age – возраст клиента
• Gender – пол клиента
• Marital Status – семейное положение клиента
• Occupation – род занятий клиента
• Monthly Income – ежемесячный доход клиента
• Educational Qualifications – образовательная квалификация клиента
• Family size – размер семьи клиент
Информация о местоположении:
• latitude – широта местонахождения клиента
• longitude – долгота местонахождения клиента
• Pin code – пин-код местонахождения клиента
Подробности заказа:
• Output – текущий статус заказа (в ожидании, подтвержден, доставлен)

• Feedback – отзыв, предоставленный клиентом после получения заказа (целевая переменная)
Набор данных `onlinefoods.csv` описывает клиентов сервиса онлайн-заказа еды: демографию, доход, размер семьи и отзывы о заказах. Цель ноутбука — подключить датасет, быстро осмотреть структуру и визуально изучить распределения ключевых признаков и их взаимосвязи без лишних действий.


```python
print("Первые 5 строк набора данных:")
display(df.head())

print("\nИнформация о столбцах:")
df.info()

print("\nЧисловые признаки:")
display(df.describe(include='number').T)

print("\nКатегориальные признаки:")
display(df.describe(include='object').T)

print("\nПропуски по столбцам:")
display(df.isna().sum())
```

```python
summary_cols = ['Output', 'Feedback', 'Gender', 'Marital_Status', 'Occupation', 'Monthly_Income']
for col in summary_cols:
    print(f"\nРаспределение {col}:")
    display(df[col].value_counts())
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='Age', bins=15, kde=True)
plt.title('Распределение возраста клиентов')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='Family_size', discrete=True)
plt.title('Размер семьи')
plt.xlabel('Число членов семьи')
plt.ylabel('Количество клиентов')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.countplot(data=df, x='Gender', ax=axes[0])
axes[0].set_title('Гендер')
axes[0].set_xlabel('')

sns.countplot(data=df, x='Marital_Status', ax=axes[1])
axes[1].set_title('Семейное положение')
axes[1].set_xlabel('')
axes[1].tick_params(axis='x', rotation=20)

sns.countplot(data=df, x='Output', hue='Feedback', ax=axes[2])
axes[2].set_title('Статус заказа и отзывы')
axes[2].set_xlabel('Статус заказа')
axes[2].legend(title='Feedback')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

sns.boxplot(data=df, x='Monthly_Income', y='Age', ax=axes[0])
axes[0].set_title('Возраст по группам дохода')
axes[0].set_xlabel('Доход')
axes[0].set_ylabel('Возраст')
axes[0].tick_params(axis='x', rotation=20)

sns.violinplot(data=df, x='Marital_Status', y='Family_size', inner='quartile', ax=axes[1])
axes[1].set_title('Размер семьи по семейному положению')
axes[1].set_xlabel('Семейное положение')
axes[1].set_ylabel('Размер семьи')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 6))
sns.scatterplot(
    data=df,
    x='longitude',
    y='latitude',
    hue='Output',
    size='Family_size',
    sizes=(20, 200),
    alpha=0.7
)
plt.title('Клиенты на карте города')
plt.xlabel('Долгота')
plt.ylabel('Широта')
plt.legend(title='Output', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['Age', 'Family_size', 'latitude', 'longitude']
plt.figure(figsize=(6, 4))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='crest', fmt='.2f')
plt.title('Корреляции числовых признаков')
plt.tight_layout()
plt.show()

sns.pairplot(df[numeric_cols + ['Monthly_Income']].assign(Monthly_Income=df['Monthly_Income'].astype(str)), diag_kind='hist', corner=True)
plt.suptitle('Парные проекции числовых признаков', y=1.02)
plt.show()
```

## Выводы

* Основные клиенты сервиса — молодые одинокие студенты с небольшим доходом или без него, семьи обычно состоят из 2–5 человек.
* Большинство заказов уже подтверждено или доставлено (статус `Yes`), отзывы преимущественно положительные.
* На карте виден охват нескольких районов: координаты концентрируются вокруг центра города, но есть и отдалённые точки.
* Распределения возраста и размеров семьи симметричны, явных выбросов немного. Группы дохода связаны с более широким возрастным диапазоном.
* Корреляции между числовыми признаками невысоки, поэтому для дальнейшего моделирования стоит уделить больше внимания категориальным переменным и отзывам.
